```python
marks = [35, 48, 52, 67, 76, 89, 95]
```

```python
bins = [0, 40, 60, 80, 100]
labels = ["Fail", "Pass", "Good", "Excellent"]

grades = pd.cut(marks, bins=bins, labels=labels)
print(grades)
```

```python
sales = [100, 150, 200, 250, 300]
df = pd.DataFrame({"Sales": sales})
print(df)
```

```python
df["Running_Avg"] = df["Sales"].expanding().mean()
print(df)
df["Running_sum"] = df["Sales"].expanding().sum()
print(df)
```

```python
df1 = pd.DataFrame({"Sales": [100, 200]})
df2 = pd.DataFrame({"Sales": [300, 400]})

# Concat
df = pd.concat([df1, df2], ignore_index=True)

# Running Average
df["Running_Avg"] = df["Sales"].expanding().mean()

# Categorize sales
df["Category"] = pd.cut(
    df["Sales"],
    bins=[0, 200, 350, 500],
    labels=["Low", "Medium", "High"]
)

print(df)
```
